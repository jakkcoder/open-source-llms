# Hugging Face token check

Loads **`HF_TOKEN`** from **`.env`** at the repo root and verifies it with `huggingface_hub.whoami`.

Put this in `.env` (do not commit):

```
HF_TOKEN=hf_...
```


````python
from __future__ import annotations

import os
from pathlib import Path

from dotenv import load_dotenv
from huggingface_hub import whoami
from huggingface_hub.utils import HfHubHTTPError

# Repo root = parent of notebooks/
ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

load_dotenv(ROOT / ".env")

token = (os.getenv("HF_TOKEN") or "").strip()
if not token:
    raise RuntimeError(
        "HF_TOKEN is missing. Add it to .env at the repo root, e.g. HF_TOKEN=hf_..."
    )

try:
    info = whoami(token=token)
except HfHubHTTPError as e:
    raise RuntimeError(f"Hub rejected the token ({e.response.status_code}): {e}") from e

safe = {
    "type": info.get("type"),
    "name": info.get("name"),
    "fullname": info.get("fullname"),
    "id": info.get("id"),
}
print("Token OK — Hub identity:")
for k, v in safe.items():
    print(f"  {k}: {v}")
````
